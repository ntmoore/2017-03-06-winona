# make life easier with a function


```python
def print_egg_label(egg_mass):
    "this function bins eggs by mass"

    if(egg_mass<50):
        #error, egg is bad, or scale is broken?
        return "error"
    
    #egg sizing machinery
    if(egg_mass>=85):
        print("jumbo")
    elif(egg_mass>=70):
        print("large")
    elif(egg_mass<70 and egg_mass>=55):
        print("medium")
    else:
        print("small")
    return "good measurement"
```
